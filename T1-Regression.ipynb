3) Do cars produced in the recent years cost more than car produced years before?

```python
# let us study how the selling price of cars has changed over the years using the manufacturing date of vehicles
# we will be using average selling price for our analysis
def condition(x):
    if 1995<=x<2000:
        return '1995-2000'
    elif 2000<=x<2005:
        return '2000-2005'
    elif 2005<=x<2010:
        return '2005-2010'
    elif 2010<=x<2015:
        return '2010-2015'
    elif 2015<=x<=2020:
        return '2015-2020'
        
df_1['Manufactured_Year_range']=df_1['year'].apply(lambda x:condition(x))    
new  = df_1.groupby('Manufactured_Year_range')['selling_price'].mean().reset_index()
new=new.rename(columns={'selling_price':'Average Selling Price'})
new
# we can see a big increase in the average selling price of vehicles when comparing the yearly interval of '2010-2015' and '2015-2020'
```

```python
# let us visualize the above dataset in a barplot
plt.figure(figsize=(8, 6))
plt.bar(new['Manufactured_Year_range'], new['Average Selling Price'], color='skyblue')
plt.xlabel('Year Of Manufacture')
plt.ylabel('Average Selling Price (Indian Rupees)')
plt.title('Average Selling Price Over The Years')

# we can see that the average selling price of vehicles has increased over time and this can be due to many factors such as inflation, tech advancements , increase in demand etc..
```

Model Building

```python
feature_variables=df.drop(['selling_price'],axis=1) # selecting our feature variable
target = df['selling_price']                         # selecting our target variable
```

```python
# splitting the dataset into  training and test data
from sklearn.model_selection import train_test_split  # importing required libraries and function 
x_train, x_test, y_train, y_test = train_test_split(feature_variables, 
                                                    target, 
                                                    test_size=0.2,  # the dataset is split into 80% training data and 20% test data
                                                    random_state=16)  # asigning a random to replicate results
```

```python
# exploring all training and test data
print(x_train.head())
print(x_test.head())    
print(y_train.head())
print(y_test.head())
```

```python
# let us now look at the shape  of all our train and test data sets
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)  # we can see the y_train and y_test have no columns,this  is because they are representing the target variable 
print(y_test.shape)   
```

```python
df.head(0)
```

```python
#let us scale the dataset
#scaling the dataset before fitting a model ensures that all features are on a comparable scale and contribute equally to the learning process
#first scale on the X_train dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaler=scaler.fit_transform(x_train)
x_test_scale=scaler.transform(x_test) #ensuring that both the training and test data are transformed in the same way to maintain consistency .
X_scaler[0] # getting the first row of our scaled x_train data
```

```python
#  Multiple linear regression
from sklearn.linear_model import LinearRegression
MLR_model=LinearRegression()
MLR_model.fit(x_train,y_train)
```

```python
# assumptions under multiple linear regression
# as the relationship between the independent and dependent variables is linear
#observations are independent of each other 
#the variance of the errors is constant across all levels of the independent variables 
#the residuals (errors) are normally distributed 
#the independent variables are not highly correlated with each other.
```

```python
# predictions on the testing data
y_prediction=MLR_model.predict(x_test)
y_prediction[0:50] # exploring predictions
```

```python
plt.figure(figsize=(5,5))
sns.displot(y_prediction, kde=True, color='blue')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()


# now  let us plot the distribution of the residuals, that is the difference between the true value of selling price and the predicted selling price
plt.figure(figsize=(5,5))
residual_value = y_test-y_prediction
sns.distplot(residual_value)
plt.xlabel('Residual Value')
plt.ylabel('Count')
plt.show()
print('The error terms (residuals) appear to be normally distributed (approximately)')
```

```python
# now we analyze if the the error term is homoscedastic as per assumtion for multiple linear regression
# the assumption holds
sns.scatterplot(x=y_prediction, y=residual_value)
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.title('Residual vs Fitted Plot')
plt.axhline(y=0, color='red', linestyle='--')  #adding a horizontal line at y=0 for reference
plt.show()
```

```python
# import the required libraries
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedStratifiedKFold
```

```python
#cross validate the data using RepeatedStratifiedKFold
#n_splits refers to the number of subsets the data will be split into for each fold
#n_repeats refers to the number of times the cross validation process takes place
cross = RepeatedStratifiedKFold(n_splits=10, n_repeats=2, random_state=123)
```

```python
parameter_dictionary = {
    'maximum_depth': [None,1,2,3],
    'criterion': ['mse', 'mae',  'rmse'],
    'maximum_features': ['sqrt', 'log2', 'auto'],
    'minimum_samples_leaf': [1,2,3,4,5,6],
    'minimum_samples_split':[1,2,3,4,5],}
#'minimum_samples_leaf' shows the minimum number of samples for a leaf node in the decision tree
#'minimum_samples_split' shows the minimum number of samples required to split an internal node

#minimum_samples_leaf and minimum_samples_split act as regularization parameters that constrain the growth of decision trees while reducing the risk of overfitting
```

```python
Grid_Search= GridSearchCV(XGBRegressor(),
                          param_grid=parameter_dictionary,
                          scoring='accuracy',
                          cv=cross,
                          n_jobs=-1,
                          verbose=10)
```

```python
Grid_Search.fit(x_train,y_train)
```

```python
Grid_Search.best_params_  # getting the best set of hyperparameters found during the grid search process
```

```python
best_grid_estimator=Grid_Search.best_estimator_
best_grid_estimator # best estimator from the grid search process
```

```python
# Make predictions using the best estimator
y_prediction = Grid_Search.best_estimator_.predict(x_test)
y_prediction[0:50]  # exploring predictions
```

```python
# distribution plot of the predictions made
sns.displot(y_prediction, kde=True, color='blue')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()



# now  let us plot the distribution of the residuals, that is the difference between the true value of selling price and the predicted selling price
residual_value = y_test-y_prediction
sns.distplot(residual_value)
```

```python
# Mean Absolute Error (MAE)
# lower MAE the better the model
# measures the average of the difference between actual values and predicted values
from sklearn.metrics import mean_absolute_error
MAE= mean_absolute_error(y_test,y_prediction)
print('Mean Absolute Error:',MAE)

# Mean Squared Error (MSE)
#measures the average squared difference between actual and predicted values
# the lower the MSE the better
from sklearn.metrics import mean_squared_error
MSE= mean_squared_error(y_test,y_prediction)
print('Mean Squared Error:',MSE)

# Root Mean Squared Error (RMSE)
#RMSE quantifies prediction error in the same measurement units as the target variable
#this makes it easier to interpret the accuracy of the model's predictions
RMSE=np.sqrt(MSE)
print('Root Mean Squared Error:',RMSE)

# R^2 Score ( Goodness Of Fit)
# shows the variation in the dependent variable (selling_price) due to variation in the independent variables ( feature variables)
from sklearn.metrics import r2_score
R2=r2_score(y_test,y_prediction)
print('R_Squared(R2 score):',R2)
```
